## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
#Data Wrangling for Play Store Data(df1)
#fixing null values in "Ratings" Column in Play Store Data (df1)
plt.figure(figsize=(5,2))
sns.histplot(df1 , x = 'Rating' , kde = True)
````

````python
df1.loc[:,'Rating'].skew() #moderately right skewed (median to be used)
````

````python
#replacing null values with median
df1.loc[:,'Rating'] = df1.loc[:,'Rating'].fillna(df1.loc[:,'Rating'].median())
df1.loc[:,'Rating'].isnull().sum()
````

````python
#converting Installs from object data type to float data type
df1.loc[:,'Installs'] = df1.loc[:, 'Installs'].astype(str)
df1.loc[:,'Installs'] = (
    df1.loc[:,'Installs']
      .str.replace('+', '', regex=False)
      .str.replace(',', '', regex=False)
      .str.strip())
df1['Installs'] = pd.to_numeric(df1['Installs'], errors='coerce')
df1.loc[:,'Installs'].dtype
````

````python
#Converting Price to float data type
df1.loc[:,'Price'] = df1.loc[:,'Price'].astype(str)
df1.loc[:,'Price'] = df1.loc[:,'Price'].str.replace('$','',regex = False)
df1['Price'] = pd.to_numeric(df1['Price'], errors='coerce')
df1.loc[:,'Price'].dtype
````

````python
df1.drop(['Size' , 'Last Updated' , 'Current Ver' , 'Android Ver'] , axis = 1 , inplace = True)
df1.dropna(inplace = True)
df1
````

````python
#Data Wrangling for User Reviews (df2)
df2 = df2.dropna(subset=['Translated_Review'])
df2.reset_index(drop=True, inplace=True)
df2
````

````python
#merging the dataset
#before merging few changes are to be made to df2 to make sure a clean join can be achieved
df2_cleaned = (
    df2
    .groupby('App')
    .agg(
        avg_sentiment_polarity=('Sentiment_Polarity', 'mean'),
        avg_sentiment_subjectivity=('Sentiment_Subjectivity', 'mean'),
        review_count=('Translated_Review', 'count')
    )
    .reset_index() #aggregate review-level sentiment metrics to the app level
)
df2_cleaned
````

````python
#merged dataset
df1.loc[:,'App'] = df1.loc[:,'App'].str.strip()
df2_cleaned.loc[:,'App'] = df2_cleaned.loc[:,'App'].str.strip()
merged_df = pd.merge(df1, df2_cleaned, on='App', how='left')
merged_df
````

````python
sentiment_df = merged_df.dropna(subset=['avg_sentiment_polarity']).reset_index(drop = True)
sentiment_df
````

````python
corr_df = merged_df[['Rating','Reviews','Installs','avg_sentiment_polarity','avg_sentiment_subjectivity','review_count']]
corr_df = corr_df.dropna(subset=['avg_sentiment_polarity'])
corr_df['log_installs'] = np.log1p(corr_df['Installs'])
corr_df['log_review_count'] = np.log1p(corr_df['review_count'])
corr_df.reset_index(drop = True)
````

### What all manipulations have you done and insights you found?

Firstly, I started by cleaning play store dataset (df1) in which I first fixed the null values inside the ratings columns. I started by finding the skewness of the columns and since the skewness was moderately right, I decided to use the median values to fill the null values of the ratings columns.

After this, I looked for columns that did not have the correct data types hence I converted price and installs from object datatype to float datatype. I also decided to drop irrelevant columns that will not be required for this analysis.

Moving on to user reviews data (df2), instead of taking each translated review, I decided to change the column to a review count column as it will be more useful for the analysis and also easier to perform join with df1. I also had to change the sentiment polarity and sentiment subjectivity columns to match the review count column.

After cleaning both the datasets, I merged the dataset with app being the common column on which the join was performed. There were also seperate df created for the correlation graph.
